```python
import json

# Define crop values as a list of (crop_w_start, crop_h_end) pairs
crop_values = [
    (1320, 3127),  
    (2712, 2902),  
    (1260, 3297),  
    (1496, 3622),  
    (1892, 3491),  
    (1177, 3006),  
    (1193, 3117),  
    (2261, 3326),  
    (1270, 3148),  
    (1403, 2869),  
    (1480, 3260),  
    (1056, 2967),  
    (214, 4233),  
    (1683, 3061),  
    (1700, 3149),  
    (2003, 2835),  
    (3338, 3457),  
    (2288, 2885),  
    (2112, 3831),  
    (726, 3380),  
    (1304, 3457),  
    (3537, 3270),  
    (2310, 2792),  
    (1551, 2923),  
    (2101, 3094),  
    (1342, 2968),  
    (2497, 3655),  
    (2393, 2467),  
    (1743, 2308),  
    (1578, 2946),  
    (1447, 2929),  
    (1793, 2800)  
]


# Function to update cx, cy, w, h for each frame
def update_frame(frame, crop_w_start, crop_h_end):
    # Set new width and height
    frame["w"] = 1700
    frame["h"] = 1400

    # Update cx and cy using the formula
    frame["cx"] = frame["cx"] - crop_w_start
    frame["cy"] = frame["cy"] - crop_h_end

# Read the JSON data from a file
def load_json(filename):
    with open(filename, 'r') as file:
        return json.load(file)

# Save the updated JSON data back to a file
def save_json(filename, data):
    with open(filename, 'w') as file:
        json.dump(data, file, indent=4)

# Main function
def main():
    # Set the correct file path for Windows
    filename = "/home/sharma/Downloads/stage_zopf/transforms.json"

    # Load the JSON data from the file
    data = load_json(filename)

    # Ensure we have enough crop values for all frames
    if len(crop_values) < len(data["frames"]):
        raise ValueError(f"Not enough crop values provided! Expected {len(data['frames'])}, but got {len(crop_values)}.")

    # Iterate through frames and update each one using the corresponding crop values
    for frame, (crop_w_start, crop_h_end) in zip(data["frames"], crop_values):
        update_frame(frame, crop_w_start, crop_h_end)

    # Save the updated data back to the JSON file
    save_json(filename, data)

    print("The JSON file has been updated.")

if __name__ == "__main__":
    main()
```
